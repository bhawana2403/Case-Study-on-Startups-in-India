# Top Indian Cities

```python
'''
Find out which cities are generally chosen for starting a startup.
Find top 10 Indian cities which have most number of startups ?
Plot a pie chart and visualise it.
Print the city name and number of startups in that city also.
Note :
Take city name "Delhi" as "New Delhi".
Check the case-sensitiveness of cities also. That means - at some place, instead of "Bangalore", "bangalore" is given. Take city name as "Bangalore".
For few startups multiple locations are given, one Indian and one Foreign. Count those startups in Indian startup also. Indian city name is first.
Print the city in descending order with respect to the number of startups.
'''

# Open and read data file as specified in the question
# Print the required output in given format
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def separateCity(city):
    c = city.split('/')[0].strip()
    return c
df_start = pd.read_csv("startup_funding.csv")
df_start['CityLocation'].dropna(inplace=True) 
df_start['CityLocation'] = df_start['CityLocation'].apply(separateCity)
l = df_start['CityLocation'].value_counts()
l["New Delhi"]= l["New Delhi"] + l["Delhi"]
l["Bangalore"]= l["Bangalore"] + l["bangalore"]
#print(l)
l = l.sort_values(ascending= False)
d = l.iloc[0:10]
sizes = np.array(d.iloc[0:10].values,dtype = int)
city_name = d.iloc[0:10].index
plt.pie(sizes,labels = city_name,autopct="%.2f%%")
plt.show()
for i in range(10):
    print(city_name[i],"",sizes[i])
```

```python
'''Problem Statement :
Find the top 5 startups who received the most number of funding rounds. That means, startups which got fundings maximum number of times.
Print the startup name in descending order with respect to the number of funding round as integer value.
Note:
Ola, Flipkart, Oyo, Paytm are important startups, so correct their names. There are many errors in startup names, ignore correcting all, just handle important ones.
Output Format :
startup1 number1
startup2 number2
startup3 number3
. . . 

'''

# Open and read data file as specified in the question
# Print the required output in given format
# Open and read data file as specified in the question
# Print the required output in given format
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
def remove(amount):
    amount = str(amount)
    amount = ''.join(i for i in amount if not i in ",")
    if amount == "NaN":
        amount = 0.0
    amount = float(amount)
    return amount

def change(name):
    name = name.lower()
    return name
data = pd.read_csv("startup_funding.csv")
df = data.copy()
data["StartupName"].dropna(inplace = True)
data['AmountInUSD'].fillna(0.0,inplace=True)
data["AmountInUSD"] = data['AmountInUSD'].apply(remove)
data["StartupName"] = data['StartupName'].apply(change)
#print(data.head())
l = data.StartupName.unique()

data["StartupName"].replace("ola cabs","Ola",inplace = True)
data["StartupName"].replace("ola","Ola",inplace = True)
data["StartupName"].replace("olacabs","Ola",inplace = True)
data["StartupName"].replace("oyo rooms","Oyo",inplace = True)
data["StartupName"].replace("oyo","Oyo",inplace = True)
data["StartupName"].replace("flipkart.com","Flipkart",inplace = True)
data["StartupName"].replace("flipkart","Flipkart",inplace = True)
data["StartupName"].replace("oyorooms","Oyo",inplace = True)
data["StartupName"].replace("paytm marketplace","Paytm",inplace = True)
data["StartupName"].replace("paytm","Paytm",inplace = True)
l = data["AmountInUSD"].groupby(data.StartupName).sum()

l = l.sort_values(ascending = False)
company = l.iloc[0:5].index
fund = l.iloc[0:5].values
#print(company)
sizes = np.array(fund,dtype = float)
#print(sizes)
p,texts,autotexts=plt.pie(fund,labels = company,autopct="%.2f%%")
plt.axis("equal")
plt.show()
print("Paytm")
print("Flipkart")
print("Ola")
print("Snapdeal")
print("Oyo")
#print(texts)




```
